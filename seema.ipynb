```python
import sqlite3
```

```python
conn = sqlite3.connect("food_wastage.db")
cursor = conn.cursor()
```

```python
cursor.execute('''
    CREATE TABLE IF NOT EXISTS providers (
        Provider_ID INTEGER PRIMARY KEY,
        Name TEXT,
        Type TEXT,
        Address TEXT,
        City TEXT,
        Contact TEXT
    )
''')
```

```python
for index, row in providers.iterrows():
    cursor.execute("""
        INSERT INTO providers (Provider_ID, Name, Type, Address, City, Contact)
        VALUES (?, ?, ?, ?, ?, ?)
    """, tuple(row))
```

```python
cursor.execute('''
    CREATE TABLE IF NOT EXISTS receivers (
        Receiver_ID INTEGER PRIMARY KEY,
        Name TEXT,
        Type TEXT,
        City TEXT,
        Contact TEXT
    )
''')
```

```python
for index, row in receivers.iterrows():
    cursor.execute("""
        INSERT INTO receivers (Receiver_ID, Name, Type, City, Contact)
        VALUES (?, ?, ?, ?, ?)
    """, tuple(row))
```

```python
cursor.execute('''
    CREATE TABLE IF NOT EXISTS food_listings (
        Food_ID INTEGER PRIMARY KEY,
        Food_Name TEXT,
        Quantity TEXT,
        Expiry_Date TEXT,
        Provider_ID TEXT,
        Provider_Type TEXT,
        Location TEXT,
        Food_Type TEXT,
      	Meal_Type TEXT
    )
''')
```

```python
for index, row in food_listings.iterrows():
    cursor.execute("""
        INSERT INTO food_listings (Food_ID, Food_Name, Quantity, Expiry_Date, Provider_ID, Provider_Type, Location, Food_Type, Meal_Type)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
    """, tuple(row))
```

```python
cursor.execute('''
    CREATE TABLE IF NOT EXISTS claims (
        Claim_ID INTEGER PRIMARY KEY,
        Food_ID TEXT,
        Receiver_ID TEXT,
        Status TEXT,
        Timestamp TEXT

    )
''')
```

```python
for index, row in claims.iterrows():
    cursor.execute("""
        INSERT INTO claims (claim_ID, Food_ID, Receiver_ID, Status, Timestamp)
        VALUES (?, ?, ?, ?, ?)
    """, tuple(row))
```

1. How many food providers and recivers are there in each city?

```python
query= ("""
    SELECT City, COUNT(*) AS Total,'Providers' AS Type
    FROM providers
    GROUP BY City

    UNION ALL

    SELECT City, COUNT(*) AS Total,'Receivers' AS Type
    FROM receivers
    GROUP BY City;
""")

df = pd.read_sql_query(query, conn)
df

```

2. Which type of food provider (restaurant, grocery store etc) contributes the most food?

```python
query = """ 
SELECT Provider_Type, SUM(Quantity) AS Total_Quantity
FROM food_listings
GROUP BY Provider_Type
ORDER BY Total_Quantity DESC;
"""

df = pd.read_sql_query(query, conn)
df

```

3. What is the contact information of food providers in specific city?

```python
query = """
SELECT Name, Contact
FROM providers
WHERE City = 'East Sheena';
"""

df = pd.read_sql_query(query, conn)
df
```

4. Which receivers have claimed the most food?

```python
query = """
SELECT r.Name, COUNT(c.Claim_ID) AS Total_Claims
FROM claims c
JOIN receivers r ON c.Receiver_ID = r.Receiver_ID
GROUP BY r.Name
ORDER BY Total_Claims DESC;
"""

df = pd.read_sql_query(query, conn)
df
```

5. What is the total quantity of food available from all providers?

```python
query = """
SELECT SUM(Quantity) AS Total_Quantity 
FROM food_listings;
"""

df = pd.read_sql_query(query, conn)
df
```

6. Which city has the highest number of food listings?

```python
query = """
SELECT Location AS City, COUNT(*) AS Total_Listings
FROM food_listings
GROUP BY Location
ORDER BY Total_Listings DESC;
"""

df = pd.read_sql_query(query, conn)
df
```

7. What are most commonly available food types?

```python
query = """
SELECT Food_Type, COUNT(*) AS Count
FROM food_listings
GROUP BY Food_Type
ORDER BY Count DESC;
"""

df = pd.read_sql_query(query, conn)
df
```

8. How many food claims have been made for each food item?

```python
query = """
SELECT f.Food_Name, COUNT(c.Claim_ID) AS Total_Claims
FROM claims c
JOIN food_listings f ON c.Food_ID = f.Food_ID
ORDER BY Total_Claims DESC;
"""

df = pd.read_sql_query(query, conn)
df
```

9. Which provider has had the highest number of successful food claims?

```python
query = """
SELECT p.Name, COUNT(c.Claim_ID) AS Successful_Claims
FROM claims c
JOIN food_listings f ON c.Food_ID = f.Food_ID
JOIN providers p ON f.Provider_ID = p.Provider_ID
WHERE c.Status = 'Completed'
GROUP BY p.Name
ORDER BY Successful_Claims DESC;
"""

df = pd.read_sql_query(query, conn)
df
```

10. What percentage of food claims are completed vs. pending vs. canceled?

```python
query = """
SELECT Status, COUNT(*) * 100.0 / (SELECT COUNT(*) FROM claims) AS Percentage 
FROM claims
GROUP BY Status;
"""

df = pd.read_sql_query(query, conn)
df
```

11. What is the average quantity of food claimed per receiver?

```python
query = """
SELECT r.Name, AVG(f.Quantity) AS Avg_Claimed_Quantity
FROM claims c
JOIN food_listings f ON c.Food_ID = f.Food_ID
JOIN receivers r ON c.Receiver_ID = r.Receiver_ID
GROUP BY r.Name;
"""

df = pd.read_sql_query(query, conn)
df
```

12.Which meal type (breakfast, lunch, dinner, snacks) is claimed the most?

```python
query = """
SELECT f.Meal_Type, COUNT(*) AS Claim_Count
FROM claims c
JOIN food_listings f ON c.Food_ID = f.Food_ID
GROUP BY f.Meal_Type
ORDER BY Claim_Count DESC;
"""

df = pd.read_sql_query(query, conn)
df
```

13. What is the total quantity of food donated by each provider?

```python
query = """
SELECT p.Name, SUM(f.Quantity) AS Total_Donated
FROM food_listings f
JOIN providers p ON f.Provider_ID = p.Provider_ID
GROUP BY p.Name
ORDER BY Total_Donated DESC;
"""

df = pd.read_sql_query(query, conn)
df
```

14. How many expired food items are still listed?

```python
query = """
SELECT COUNT(*) AS Expired_Items
FROM food_listings
WHERE Expiry_Date < Date('now');
"""

df = pd.read_sql_query(query, conn)
df
```

15. What are the toP 5 cities with the highest completed claims?

```python
query = """
SELECT f.Location AS City, COUNT(*) AS Completed_Claims
FROM claims c
JOIN food_listings f ON c.Food_ID = f.Food_ID
WHERE c.Status ='Completed'
GROUP BY f.Location
ORDER BY Completed_Claims DESC
LIMIT 5;
"""

df = pd.read_sql_query(query, conn)
df
```

16. What are the top 5 most donated food items?

```python
query = """
SELECT Food_Name, SUM(Quantity) AS Total_Donated
FROM food_listings
GROUP BY Food_Name
ORDER BY Total_Donated DESC
LIMIT 5;
"""

df = pd.read_sql_query(query, conn)
df
```

17. How many inactive providers?

```python
query = """
SELECT p.Name, p.City
FROM providers p
LEFT JOIN food_listings f ON p.Provider_ID = f.Provider_ID
WHERE f.Food_ID IS NULL;
"""

df = pd.read_sql_query(query, conn)
df
```

18. Claims trend over time

```python
query = """
SELECT DATE(Timestamp) AS Claim_Date, COUNT(*) AS Total_Claim
FROM claims
GROUP BY Claim_Date
ORDER BY Claim_Date ASC;
"""

df = pd.read_sql_query(query, conn)
df
```

19. Claims trend by city

```python
query = """
SELECT f.Location AS City, COUNT(*) AS Total_Claims
FROM claims c
JOIN food_listings f ON c.Food_ID = f.Food_ID
GROUP BY f.Location 
ORDER BY Total_Claims DESC;
"""

df = pd.read_sql_query(query, conn)
df
```

20. Claims made after food expiry 

```python
query = """
SELECT c.Claim_ID, f.Food_Name, f.Expiry_Date, c.Timestamp
FROM claims c
JOIN food_listings f ON c.Food_ID = f.Food_ID
WHERE DATE(c.Timestamp) > DATE(f.Expiry_Date);
"""

df = pd.read_sql_query(query, conn)
df
```
